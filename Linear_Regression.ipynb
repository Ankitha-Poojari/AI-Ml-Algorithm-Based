```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create dataset (Hours studied vs Exam scores)
# X must be 2D for sklearn (n_samples, n_features)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # Hours studied
y = np.array([10, 20, 30, 40, 50, 60])           # Exam scores

# Step 3: Create Linear Regression model
model = LinearRegression()

# Step 4: Train (fit) the model
model.fit(X, y)

# Step 5: Get slope (m) and intercept (c)
m = model.coef_[0]      # slope
c = model.intercept_    # intercept
print(f"Equation of line: y = {m:.2f}x + {c:.2f}")

# Step 6: Make predictions
y_pred = model.predict(X)

# Step 7: Plot data and regression line
plt.scatter(X, y, color="blue", label="Actual Data")   # actual points
plt.plot(X, y_pred, color="red", label="Regression Line")  # best-fit line
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Simple Linear Regression Example")
plt.legend()
plt.show()
```
